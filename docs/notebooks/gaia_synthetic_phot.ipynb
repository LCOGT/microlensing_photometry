# Gaia Synthetic Photometry 

Gaia DR3 catalogs include photometry directly from the Gaia mission in their rp, bp passbands.  
More recently, the [Gaia Synthetic Photometry catalog](https://gaia.aip.de/metadata/gaiadr3/synthetic_photometry_gspc/) has become available through astroquery. 

```python
from astroquery.gaia import Gaia
from astropy.coordinates import SkyCoord
from astropy import units as u
```

Following the example in astroquery's documentation, output available Gaia tables 

```python
tables = Gaia.load_tables(only_names=True)
for table in tables:
    print(table.get_qualified_name())
```

It looks like this catalog can be queried under the name "gaiadr3.gaiadr3.synthetic_photometry_gspc"

```python
# Using OGLE-2024-BLG-0034 as an example 
target = SkyCoord('17:44:10.400', '-39:06:47.60', frame='icrs', unit=(u.hourangle, u.deg))
```

```python
qjob = Gaia.cone_search_async(target, radius=u.Quantity(1.0, u.deg))
results = qjob.get_results()
```

```python
results.pprint()
```

```python
results.colnames
```

According to the documentation, this photometry is derived from objects in DR3 for which Gaia spectra were available, so this is very incomplete in crowded fields.  
For the time being, there is no clear advantage to changing from the DR3 catalog.
